```python
import face_recognition
import cv2
import numpy as np
import csv
import os
from datetime import datetime
```

```python
class main:
    def _init_(self):
        self.data()
        self.menu()
        
    
    def data(self):
        self.face_dict={'heet':face_recognition.face_encodings(face_recognition.load_image_file("E:/Vs_code/Python/project/face recognition/heet.jpg"))[0],
                   'sachin':face_recognition.face_encodings(face_recognition.load_image_file("E:/Vs_code/Python/project/face recognition/sachin.jpg"))[0],
                   'kunj':face_recognition.face_encodings(face_recognition.load_image_file("E:/Vs_code/Python/project/face recognition/kunj.jpg"))[0],
                   'sattu':face_recognition.face_encodings(face_recognition.load_image_file("E:/Vs_code/Python/project/face recognition/sattu.jpg"))[0]}
        
        self.known_face_encoding = [
        self.face_dict['heet'],
        self.face_dict['sachin'],
        self.face_dict['kunj'],
        self.face_dict['sattu'],
        ]
        
        self.known_faces_names = [
        "heet",
        "sachin",
        "kunj",
        "sattu"
        ]
       
    
    
    def menu(self):
        choice = int(input('''
        Press 1 for attendance
        Press 2 for new entry
        Press 3 for exit
        -->'''))
        
        if (choice == 1):
            self.face_recognition()
        elif(choice == 2):
            self.entry()
        else:
            print('Thank you')
    
    
    def face_recognition(self):
        video_capture = cv2.VideoCapture(0)
        students = self.known_faces_names.copy()

        face_locations = []
        face_encodings = []
        face_names = []
        s=True

        now = datetime.now()
        current_date = now.strftime("%Y-%m-%d")

        f = open(current_date+'.csv','w+',newline = '')
        lnwriter = csv.writer(f)

        while True:
            _,frame = video_capture.read()
            small_frame = cv2.resize(frame,(0,0),fx=0.25,fy=0.25)
            rgb_small_frame = small_frame[:,:,::-1]
            if s:
                face_locations = face_recognition.face_locations(rgb_small_frame)
                face_encodings = face_recognition.face_encodings(rgb_small_frame,face_locations)
                face_names = []
                for face_encoding in face_encodings:
                    matches = face_recognition.compare_faces(self.known_face_encoding,face_encoding)
                    name=""
                    face_distance = face_recognition.face_distance(self.known_face_encoding,face_encoding)
                    best_match_index = np.argmin(face_distance)
                    if matches[best_match_index]:
                        name = self.known_faces_names[best_match_index]

                    face_names.append(name)
                    if name in self.known_faces_names:
                        font = cv2.FONT_HERSHEY_SIMPLEX
                        bottomLeftCornerOfText = (10,100)
                        fontScale = 1.5
                        fontColor = (255,0,0)
                        thickness = 3
                        lineType = 2

                        cv2.putText(frame,name+' Present', 
                            bottomLeftCornerOfText, 
                            font, 
                            fontScale,
                            fontColor,
                            thickness,
                            lineType)

                        if name in students:
                            students.remove(name)
                            print(students)
                            current_time = now.strftime("%H-%M-%S")
                            lnwriter.writerow([name,current_time])
            cv2.imshow("attendence system",frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        video_capture.release()
        cv2.destroyAllWindows()
        f.close()
        self.menu()
        
    def entry(self):
        new_student = input('Enter name of the new student:')
        
        if(new_student in self.face_dict):
            print('This student is already in the list')
        
        else:
            file_name = input('Enter the name of the file of the student')
            file_path = "E:/Vs_code/Python/project/face recognition/"+file_name
            self.face_dict[new_student] = face_recognition.face_encodings(face_recognition.load_image_file(file_path))[0]
            self.known_face_encoding.append(self.face_dict[new_student])
            self.known_faces_names.append(new_student)
        
        self.menu()
```
